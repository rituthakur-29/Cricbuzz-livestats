## Stitching it all together

````python
def merge_seats(x):
    if 2 <= x <= 4:
        return '2-4'
    elif x > 5:
        return '>5'
    else:
        return '5'

def preprocess(df):
    df = df.loc[df.mileage != 0].copy()
    outlier_theshold = df.selling_price.quantile(0.95)
    df.loc[df.selling_price > outlier_theshold, 'selling_price'] = outlier_theshold
    df['age'] = pd.to_datetime('now', utc=True).date().year - df['year']
    df['full_name'] = df['full_name'].str.upper()
    df['seats'] = df['seats'].apply(merge_seats)
    df.drop(columns=['year'])
    return df

def feature_engineering(df):
    df['make'] = df.full_name.apply(lambda x : x.split()[0])
    df['model'] = df.full_name.apply(lambda x : " ".join(x.split()[1:]))
    one_hot_encode_cols = ['seller_type', 'fuel_type', 'transmission_type', 'seats']
    for column in one_hot_encode_cols:
        df = pd.concat([df, pd.get_dummies(df[column]).iloc[: , 1:]], axis=1)
    df.drop(columns=one_hot_encode_cols + ['full_name'], inplace=True)
    df['make'] = df.groupby('make')['selling_price'].transform('median')
    df['model'] = df.groupby('model')['selling_price'].transform('median')
    scaler = MinMaxScaler()
    df = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)
    return df
````

````python
df= pd.read_csv("train-cars24-car-price.csv")
df= preprocess(df)
df= feature_engineering(df)
df.head()
````

# Insights

### 1. Price depreciation over time

````python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
sns.lineplot(x='age', y='selling_price', data=df, estimator='mean')
plt.title('Price Depreciation over Time')
plt.xlabel('Car Age')
plt.ylabel('Average Selling Price')
plt.show()
````

### 2. High mileage vs low mileage comparison

````python
df['age_group'] = pd.cut(df['age'], bins=3, labels=['New', 'Mid', 'Old'])

sns.boxplot(x='km_category', y='selling_price', hue='age_group', data=df)
````

## 3. Point of maximum value loss

````python
depreciation = df.groupby('age')['selling_price'].mean().reset_index()
depreciation = depreciation.sort_values('age')
depreciation['price_drop'] = depreciation['selling_price'].diff()

plt.figure(figsize=(10,5))
sns.lineplot(x='age', y='price_drop', data=depreciation, marker='o')
plt.title('Price Drop by Age')
plt.ylabel('Drop in Price')
plt.show()
````

## 4. Engine & Power Influence

````python
plt.figure(figsize=(8,5))
sns.scatterplot(x='engine', y='selling_price', data=df)
plt.title('Engine vs Price')
plt.show()

plt.figure(figsize=(8,5))
sns.scatterplot(x='max_power', y='selling_price', data=df)
plt.title('Max Power vs Price')
plt.show()
````

## 5. Fuel Type Impact

````python
fuel_cols = ['Diesel', 'Petrol', 'Electric', 'LPG']

fuel_data = []
for col in fuel_cols:
    temp = df[df[col] == 1]
    avg_price = temp['selling_price'].mean()
    fuel_data.append([col, avg_price])

fuel_df = pd.DataFrame(fuel_data, columns=['Fuel Type', 'Avg Price'])

plt.figure(figsize=(8,5))
sns.barplot(x='Fuel Type', y='Avg Price', data=fuel_df)
plt.title('Price by Fuel Type')
plt.show()
````
